```python
so_mapping_config = read_configs(config_type="contracts", name="so.mapping.json")
so_mapping_config
```

```python
customer_mapping = pd.read_csv(Path(raw_path)/"CustomerMapping.csv")
customer_mapping
```

```python
customer_mapping["Customer Type"] = customer_mapping["Customer Type"].str.capitalize()
customer_mapping["Customer Type"] = customer_mapping["Customer Type"].replace({"Marketing deal":"Deal-Marketing", "Grow deal":"Deal-Growth"})
customer_mapping = customer_mapping.fillna({"Customer Type":"Unknown"})
customer_mapping["Customer Type"].value_counts()
```

```python
mapping = customer_mapping.set_index("Customer")["Customer Type"].to_dict()
mapping
```

```python
so_mapping_config["customer_mapping"] = mapping 
so_mapping_config
```
